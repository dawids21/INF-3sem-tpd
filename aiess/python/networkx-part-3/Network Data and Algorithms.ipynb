```python
edges = []
n = 10

for u in range(n):
    for v in range(n):
        if u % 3 == 0 and u > v:
            edges.append((u, v))
        elif u % 5 == 0 and u < v:
            edges.append((u, v))
```

### Greedy modularity 

The `greedy_modularity_communities()` function in NetworkX employs a greedy approach to find communities in a graph by iteratively optimizing the modularity score. It begins by considering each node as a separate community. Then, in each step, it evaluates merging every possible pair of existing communities and calculates the resulting change in modularity. The merge that yields the largest positive increase in modularity (or the smallest decrease if no positive increase is possible) is performed. This process continues until no further merges can increase the modularity score. While computationally efficient, this greedy approach doesn't guarantee finding the globally optimal community structure, as it can get stuck in local optima.

The modularity score $Q$ is defined by the following formula for a given partition of the graph into communities:

$$Q = \frac{1}{2m} \sum_{i,j} \left( A_{ij} - \frac{k_i k_j}{2m} \right) \delta(c_i, c_j)$$

Where:

* $A_{ij}$ is the adjacency matrix of the graph (1 if there's an edge between nodes $i$ and $j$, 0 otherwise).
* $k_i$ is the degree of node $i$.
* $k_j$ is the degree of node $j$.
* $m$ is the total number of edges in the graph.
* $c_i$ is the community to which node $i$ belongs.
* $c_j$ is the community to which node $j$ belongs.
* $\delta(c_i, c_j)$ is the Kronecker delta function, which is 1 if nodes $i$ and $j$ belong to the same community ($c_i = c_j$), and 0 otherwise.

In simpler terms, the formula calculates the fraction of edges within communities minus the expected fraction of such edges if the graph were randomly rewired while preserving the degree of each node. The greedy algorithm iteratively merges communities to find a partition that yields a high value of $Q$.

```python
communities = nx.community.greedy_modularity_communities(Fs)
print(f"Number of communities: {len(communities)}")
```

The object returned by the `greedy_modularity_communities()` function is a list of frozensets, where each frozenset contains the nodes assigned to a given community. We have to map these communities to particular nodes in the graph.

```python
from typing import List, FrozenSet

def find_community(node_id: int, communities: List[FrozenSet[int]]) -> int:
    for community_id, community_members in enumerate(communities):
        if node_id in community_members:
            return community_id
    return -1
```

### Girvan-Newman Method

The Girvan–Newman algorithm identifies community structures within a network through an iterative edge removal process. Specifically, the algorithm operates by progressively eliminating edges from the network's topology. The resultant disconnected components that emerge following these removals are then interpreted as the inherent community divisions within the network. A key characteristic of this approach is its focus on identifying and removing edges that are hypothesized to reside *between* cohesive groups of nodes (communities), rather than attempting to quantify the centrality of edges *within* those groups.

```python
_communities = nx.community.girvan_newman(Fs)
communities = next(_communities)
print(f"Number of communities: {len(communities)}")
```

## Beam search

Beam search is a search algorithm that explores a graph or network by expanding the most promising nodes in a limited set. It's an optimization technique often used in problems where finding the absolute best path or solution is computationally expensive.

Here's a breakdown of how it generally works:

1.  **Initialization:** The search starts at a given node.
2.  **Expansion:** Instead of exploring all possible next steps, beam search only considers the "beam width" number of most likely or highest-scoring next nodes. A heuristic or scoring function determines these "most promising" nodes.
3.  **Beam:** These selected nodes form the "beam" – the set of nodes to be further explored.
4.  **Iteration:** The process repeats: for each node in the beam, the algorithm generates its successors, scores them, and again selects the top "beam width" number of nodes to form the new beam.
5.  **Termination:** The search continues until a goal is reached or a maximum depth is exceeded.

In essence, beam search is a trade-off. It sacrifices completeness (the guarantee of finding the optimal solution) for efficiency, by focusing on a limited set of the most promising paths. We will illustrate the beam search idea for searching for nodes with high centrality measures.

```python
import math

def progressive_widening_search(G, source, value, condition, initial_width=1):
    """Progressive widening beam search to find a node.

    The progressive widening beam search involves a repeated beam
    search, starting with a small beam width then extending to
    progressively larger beam widths if the target node is not
    found. This implementation simply returns the first node found that
    matches the termination condition.

    `G` is a NetworkX graph.

    `source` is a node in the graph. The search for the node of interest
    begins here and extends only to those nodes in the (weakly)
    connected component of this node.

    `value` is a function that returns a real number indicating how good
    a potential neighbor node is when deciding which neighbor nodes to
    enqueue in the breadth-first search. Only the best nodes within the
    current beam width will be enqueued at each step.

    `condition` is the termination condition for the search. This is a
    function that takes a node as input and return a Boolean indicating
    whether the node is the target. If no node matches the termination
    condition, this function raises :exc:`NodeNotFound`.

    `initial_width` is the starting beam width for the beam search (the
    default is one). If no node matching the `condition` is found with
    this beam width, the beam search is restarted from the `source` node
    with a beam width that is twice as large (so the beam width
    increases exponentially). The search terminates after the beam width
    exceeds the number of nodes in the graph.

    """

    if condition(source):
        return source

    log_m = math.ceil(math.log2(len(G)))
    for i in range(log_m):
        width = initial_width * pow(2, i)

        for u, v in nx.bfs_beam_edges(G, source, value, width):
            if condition(v):
                return v

    raise nx.NodeNotFound("No node satisfied the termination condition")
```

```python
centrality = nx.eigenvector_centrality(F)
avg_centrality = sum(centrality.values()) / len(F)
max_centrality = max(centrality.values())
condition = lambda v: centrality[v] > 0.99 * max_centrality
value = centrality.get
source = list(F.nodes())[0]

best_node = progressive_widening_search(F, source, value, condition)
print(f"Found node {best_node} with centrality {centrality[best_node]}")
```

## Graph coloring

# Network resilience

Network robustness and resilience are key concepts for understanding how networks respond to stress, like when nodes fail or are attacked. Robustness is a network's ability to maintain its structure, while resilience is its ability to recover from failures. Network robustness can be assessed by observing how the network's structure changes when nodes or edges are removed. This involves measuring metrics such as the size of the largest connected component or the average shortest path length before and after removals. We will simulate the removal of random nodes as well as targeted attacks on a network and visualize how these removals impact the network's structure.

```python
def largest_component_size(G):
    return len(max(nx.connected_components(G), key=len))

def simulate_random_failures(G, num_failures):
    G_copy = G.copy()
    nodes_to_remove = np.random.choice(list(G_copy.nodes), num_failures)
    G_copy.remove_nodes_from(nodes_to_remove)
    return G_copy, nodes_to_remove

def simulate_targeted_attacks(G, num_attacks):
    G_copy = G.copy()
    high_degree_nodes = sorted(G_copy.degree, key=lambda x: x[1], reverse=True)[:num_attacks]
    G_copy.remove_nodes_from([node for node, _ in high_degree_nodes])
    return G_copy, [node for node, _ in high_degree_nodes]
```

```python
num_failures = 10
num_attacks  = 10

G = nx.barabasi_albert_graph(n=100, m=2)

# simulate random failures
G_random_failed, removed_random = simulate_random_failures(G, num_failures)
random_failed_size = largest_component_size(G_random_failed)

# simulate targeted attacks
G_targeted_attacked, removed_targeted = simulate_targeted_attacks(G, num_attacks)
targeted_attacked_size = largest_component_size(G_targeted_attacked)

print(f"Initial largest component size: {largest_component_size(G)}")
print(f"Largest component size after {num_failures} random failures: {random_failed_size}")
print(f"Largest component size after {num_attacks} targeted attacks: {targeted_attacked_size}")
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 3, 1)
nx.draw(G, 
        node_color='skyblue', 
        node_size=50, 
        edge_color='gray',
        alpha=0.8,
       )
plt.title(f"Original graph")

plt.subplot(1, 3, 2)
nx.draw(G_random_failed, 
        node_color='skyblue', 
        node_size=50, 
        edge_color='gray',
        alpha=0.8,
       )
plt.title(f"Graph after {num_failures} random failures")

plt.subplot(1, 3, 3)
nx.draw(G_targeted_attacked, 
        node_color='skyblue', 
        node_size=50, 
        edge_color='gray',
        alpha=0.8
       )
plt.title(f"Graph after {num_attacks} targeted attacks")
plt.show()
```

The tensor product is a way of multiplying vectors (or tensors, which are generalizations of vectors and matrices to higher dimensions) to produce a new tensor.  It's different from element-wise multiplication or dot products. The key idea is that it combines the components of the original vectors in a specific way to create a larger vector or tensor. If you have a vector of shape `(m,)` and a vector of shape `(n,)`, the tensor product will have a shape of `(m*n,)`.

The tensor product of two matrices, A (with dimensions m x n) and B (with dimensions p x q), results in a matrix of size (m\*p) x (n\*q).  Each element of matrix A is multiplied by the entire matrix B.

For example, if

$A = \begin{bmatrix} a_{11} & a_{12} \\ a_{21} & a_{22} \end{bmatrix}$

and

$B = \begin{bmatrix} b_{11} & b_{12} \\ b_{21} & b_{22} \end{bmatrix}$

then the tensor product A ⊗ B is:

$A \otimes B = \begin{bmatrix}
a_{11} \begin{bmatrix} b_{11} & b_{12} \\ b_{21} & b_{22} \end{bmatrix} &
a_{12} \begin{bmatrix} b_{11} & b_{12} \\ b_{21} & b_{22} \end{bmatrix} \\
a_{21} \begin{bmatrix} b_{11} & b_{12} \\ b_{21} & b_{22} \end{bmatrix} &
a_{22} \begin{bmatrix} b_{11} & b_{12} \\ b_{21} & b_{22} \end{bmatrix}
\end{bmatrix}$

Expanding this, we get:

$A \otimes B = \begin{bmatrix}
a_{11}b_{11} & a_{11}b_{12} & a_{12}b_{11} & a_{12}b_{12} \\
a_{11}b_{21} & a_{11}b_{22} & a_{12}b_{21} & a_{12}b_{22} \\
a_{21}b_{11} & a_{21}b_{12} & a_{22}b_{11} & a_{22}b_{12} \\
a_{21}b_{21} & a_{21}b_{22} & a_{22}b_{21} & a_{22}b_{22}
\end{bmatrix}$

Let's write a function that implements the [Kronecker Graph of order k](http://arxiv.org/pdf/0812.4905v2.pdf). This is simply the `tensor_product` of a graph applied to itself k times. That is, if we have a graph $G$, with adjacency matrix $A$ and kronecker product(tensor product) $\otimes$, the _Kronecker Graph_ of order $k$ is a graph with Adjacency matrix

$$(A)^k = \underbrace{A \otimes A \otimes \cdots \otimes A}_{ k \text{ terms}}$$
